```python
from datetime import datetime as d
ngay = d.strptime('2020-12-4', '%Y-%m-%d')

def toTime(x):
    return d.strptime(x, '%Y-%m-%d')
dataframe['Date_reported'].apply(toTime)
```

```python
# Grouping rows by time 
new_df = dataframe.set_index(dataframe['Date_reported'].apply(toTime))
new_df.head(2)
```

```python
# Group by year
new_df.resample('y').sum()
```

```python
# Group by month
new_df.resample('m').sum().head(2)
```

```python
# Applying a function to groups
# Calculate total New_cases for each year
new_df.groupby('Country_code').resample('y').sum().head(10)
```

```python
df_sales = pd.DataFrame()
df_sales['Department'] = [random.choice(['A', 'B', 'C']) for _ in range(20)]
df_sales['TV'] = [random.randint(1, 10) for _ in range(20)]
df_sales['AC'] = [random.randint(1, 10) for _ in range(20)]
df_sales['Fan'] = [random.randint(1, 10) for _ in range(20)]
df_sales.head(2)
```

```python
# Find the total number of TVs, ACs, Fans were saled by each department
df_sales.groupby('Department').sum()
```

```python
# Count the number of rows in each group
df_sales.groupby('Department').apply(lambda x: x.count())
```

```python
# Increment 1 for column TV and AC
df_sales.groupby('Department')['TV', 'AC'].apply(lambda x: x + 1).head(3)
```

```python
df_sales.loc[: , ['TV', 'AC']].apply(lambda x: x + 1).head(3)
```

```python
# Concatenating dataframes
df_1 = pd.DataFrame()
df_1['id'] = [i for i in range(1, 4)]
df_1['Name'] = ['A', 'B', 'C']
df_1
```

```python
df_2 = pd.DataFrame()
df_2['id'] = [i for i in range(4, 8)]
df_2['Name'] = ['M', 'K', 'H', 'I']
df_2
```

```python
# Concate along the row axis (axis = 0)
pd.concat([df_1, df_2], axis=0)
```

```python
# Concate along the column axis (axis = 1)
pd.concat([df_1, df_2], axis=1)
```

```python
# Merging the dataframe
df_3 = pd.DataFrame()
df_3['id'] = [i for i in range(1, 4)] + [7]
df_3['Name'] = ['A', 'B', 'C', 'K']
df_3
```

```python
df_4 = pd.DataFrame()
df_4['id'] = [i for i in range(1, 6)]
df_4['Sale'] = [random.randint(20, 50) for _ in range(5)]
df_4
```

```python
pd.merge(df_3, df_4, on='id', how='right')
```

```python
pd.merge(df_3, df_4, on='id', how='left')
```

```python
pd.merge(df_3, df_4, on='id', how='inner')
```

```python
pd.merge(df_3, df_4, on='id', how='outer')
```

```python
pd.merge(df_3, df_4, on='id')
```
