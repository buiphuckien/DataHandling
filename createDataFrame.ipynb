# How To Use Pandas To Create A DataFrame

## Create a data frame

```python
# Import library
import pandas as pd
import random

# Suppress warning in python
import warnings
warnings.filterwarnings("ignore")

# Create data frame
df = pd.DataFrame()

# Create column name 
df['Name'] = ['John', 'Tri', 'Mary', 'Tom', 'Anne']
df['Math'] = [random.randint(0, 10) for i in range(5)]
df['Physics'] = [random.randint(0, 10) for i in range(5)]
df['Chemistry'] = [random.randint(0, 10) for i in range(5)]
df['Literature'] = [random.randint(0, 10) for i in range(5)]
df
```

## Add rows to the data frame

```python
# Add one row
index=['Name', 'Math', 'Physics', 'Chemistry', 'Literature']
new_row = pd.Series(['Han', 5, 6, 7, 9], index=index)
df = df.append(new_row, ignore_index = True)

# Add many rows
rows = [pd.Series(['Na', 5, 6, 2, 3], index=df.columns),
       pd.Series(['Nam', 9, 6, 6, 4], index=df.columns),
       pd.Series(['Bao', 9, 10, 1, 2], index=df.columns)]
df = df.append(rows, ignore_index = True)
df
```
